## Adding Noise

```python
import numpy as np
gaussian_noise = np.random.normal(0, 1, shape) # Standard Normal noise

noisy_imgL = imgL + gaussian_noise
noisy_imgR = imgR + gaussian_noise

f,ax = plt.subplots(1,2, figsize=(10,5))

ax[0].imshow(noisy_imgL, cmap="gray")
ax[0].axis('off')
ax[0].set_title('Left Noisy image')

ax[1].imshow(noisy_imgR, cmap="gray")
ax[1].axis('off')
ax[1].set_title('Right Noisy image')
```

```python
noisy_imgL_gray = noisy_imgL.astype('uint8')
noisy_imgR_gray = noisy_imgR.astype('uint8')
```

## Creating StereoBM object for Disparity Maps

```python
stereo = cv2.StereoBM_create(numDisparities=128, blockSize=9)

disparity = stereo.compute(noisy_imgL_gray, noisy_imgR_gray)
plt.imshow(disparity, cmap="gray")
plt.axis('off')
```

```python
cv2.imwrite('disparity_map__original.jpg',disparity)
```

```python
OCCLUSION = 3.84

def min_cost(a, b, c):
    return min(a, b, c)

def sqr(a):
    return a * a

def generate_map(left, right):
    img_height, img_width = left.shape
    disparity_map = np.zeros((img_height, img_width), dtype=np.uint8)

    for line in range(img_height):
        cost_matrix = np.zeros((img_width + 1, img_width + 1), dtype=np.float64)

        cost_matrix[0, 0] = 0.0
        cost_matrix[1:, 0] = np.arange(1, img_width + 1) * OCCLUSION
        cost_matrix[0, 1:] = np.arange(1, img_width + 1) * OCCLUSION

        for i in range(1, img_width + 1):
            for j in range(1, img_width + 1):
                left_pixel = left[line, i - 1]
                right_pixel = right[line, j - 1]

                no_occlusion = cost_matrix[i - 1, j - 1] + (1.0 / 64) * sqr(right_pixel - left_pixel)
                i_occlusion = cost_matrix[i - 1, j] + OCCLUSION
                j_occlusion = cost_matrix[i, j - 1] + OCCLUSION

                cost_matrix[i, j] = min_cost(no_occlusion, i_occlusion, j_occlusion)

        i, j = img_width, img_width
        while i > 0 and j > 0:
            left_pixel = left[line, i - 1]
            right_pixel = right[line, j - 1]

            if cost_matrix[i, j] == cost_matrix[i - 1, j] + OCCLUSION:
                disparity_map[line, i - 1] = 0
                i -= 1
            elif cost_matrix[i, j] == cost_matrix[i, j - 1] + OCCLUSION:
                j -= 1
            else:
                disparity_map[line, i - 1] = abs(i - j)
                i -= 1
                j -= 1

    return disparity_map
```

```python
disparity_img = generate_map(noisy_imgL_gray, noisy_imgR_gray)
disparity_img = cv2.normalize(disparity_img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
```

```python
# Denoising
disparity_img = cv2.fastNlMeansDenoising(disparity_img, None, 7, 7, 21)
```

## Disparity with different parametric combinations

```python
# Define parameter ranges
block_sizes = list(range(5, 22, 2))  # Must be odd and >=5
max_disps = [16,32,48,64,128]  # Must be divisible by 16
```

```python
n_rows, n_cols
```

```python
# Set up plot grid
n_rows = len(block_sizes)
n_cols = len(max_disps)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(3*n_cols, 3*n_rows))

for i, blockSize in enumerate(block_sizes):
    for j, max_disp in enumerate(max_disps):
        if max_disp % 16 != 0 or blockSize % 2 == 0 or blockSize < 5:
            continue  # Ensure valid parameters

        # Create StereoBM object
        stereo = cv2.StereoBM_create(numDisparities=max_disp, blockSize=blockSize)
        stereo.setSpeckleWindowSize(100)
        stereo.setSpeckleRange(32)

        # Compute disparity map
        disparity = stereo.compute(imgL, imgR)
        disp_uint8 = cv2.normalize(disparity, None, 0, 255, cv2.NORM_MINMAX)
        disp_uint8 = np.uint8(disp_uint8)

        # Plot
        ax = axes[i, j]
        ax.imshow(disp_uint8, cmap='gray')
        ax.set_title(f'BS={blockSize}, MD={max_disp}')
        ax.axis('off')

plt.tight_layout()
plt.show()
```
