## Importing necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, title="Image", cmap='gray'):
    plt.figure(figsize=(5,5))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

## Reading the image

```python
# Load stereo image pair (left and right views)
imgL = cv2.imread('data/left.jpg', cv2.IMREAD_GRAYSCALE)
imgR = cv2.imread('data/right.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
imgL.dtype
```

```python
shape = imgL.shape
# NOTE: both the images have same shape as the camera planes were perfectly aligned
```

## Displaying the images

```python
if imgL is None or imgR is None:
    raise FileNotFoundError("Make sure 'left.jpg' and 'right.jpg' are in the working directory")

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(imgL, cmap="gray")
plt.title("Left Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(imgR, cmap="gray")
plt.title("Right Image")
plt.axis('off')
```

## Adding Noise

```python
import numpy as np
gaussian_noise = np.random.normal(0, 1, shape) # Standard Normal noise

noisy_imgL = imgL + gaussian_noise
noisy_imgR = imgR + gaussian_noise

f,ax = plt.subplots(1,2, figsize=(10,5))

ax[0].imshow(noisy_imgL, cmap="gray")
ax[0].axis('off')
ax[0].set_title('Left Noisy image')

ax[1].imshow(noisy_imgR, cmap="gray")
ax[1].axis('off')
ax[1].set_title('Right Noisy image')
```

```python
noisy_imgL_gray = noisy_imgL.astype('uint8')
noisy_imgR_gray = noisy_imgR.astype('uint8')
```

## Creating StereoBM object for Disparity Maps

```python
stereo = cv2.StereoBM_create(numDisparities=128, blockSize=9)

disparity = stereo.compute(noisy_imgL_gray, noisy_imgR_gray)
plt.imshow(disparity, cmap="gray")
plt.axis('off')
```

```python
cv2.imwrite('disparity_map__original.jpg',disparity)
```
